#### Stationarity
- TS data is said to be stationary if it does not exhibit trend or seasonality
- i.e. the mean, variance and covariance are constant for any segment of the series and are not functions of time

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df2 = pd.read_csv('../Data/samples.csv', index_col=0, parse_dates=True)
df2.head()
```

```python
# Stationary data
df2['a'].plot(title='STATIONARY DATA').autoscale(axis='x', tight=True);
```

```python
# Non stationary data
df2['b'].plot(title="NON STATIONARY DATA").autoscale(axis='x', tight=True);
```

```python
df2['c'].plot(title="NON STATIONARY DATA").autoscale(axis='x', tight=True);
```

#### Differencing
- Non stationary data can be made stationary through differencing
- Differencing reduces the number of rows
- diff = df['b'] - df['b'].shift(1)

```python
from statsmodels.tsa.statespace.tools import diff
```

```python
df2['d1'] = diff(df2['b'], k_diff=1)
```

```python
df2['d1'].plot(title="First Difference Data").autoscale(axis='x', tight=True);
```
