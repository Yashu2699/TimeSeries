```python
import pandas as pd
import numpy as np
%matplotlib inline
import statsmodels.api as sm

# Load a non-stationary dataset
df1 = pd.read_csv('../Data/airline_passengers.csv',index_col='Month',parse_dates=True)
df1.index.freq = 'MS'

# Load a stationary dataset
df2 = pd.read_csv('../Data/DailyTotalFemaleBirths.csv',index_col='Date',parse_dates=True)
df2.index.freq = 'D'
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

#### ACF
- Sharp dropoff indicates there is no AR component in the ARIMA model
- Blue area represents the relevance of that time step (points in the blur area are not correlated to the current time step)

```python
title = "Autocorrelation: Daily Female Births"
lags = 40
plot_acf(df2, title=title, lags=lags);
```

```python
# This is a typical ACF plot for stationary data. Sharp dropoff in the values
```

```python
title = "Autocorrelation: Airline Passengers"
plot_acf(df1, title=title, lags=40);
```

```python
# gradual dropoff indicates that this plot is non stationary
```

#### PACF

```python
title='Partial Autocorrelation: Daily Female Births'
lags=40
plot_pacf(df2,title=title,lags=lags);
```

```python
title='PACF: Airline Passengers First Difference'
lags=40
plot_pacf(df1,title=title,lags=lags);
```
